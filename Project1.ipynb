## Cleansing the data 
We will clean the data, handle the missing values, and convert the data types, amd removes the duplicates, and visialize the clean data

```python
df.head(10)
```

```python
# Removing the duplicates
df.drop_duplicates(inplace=True)
```

```python
# checking the null values 
df.isna().sum()
```

```python
# Check for missing values
missing_data = df.isnull().sum()
missing_data_percentage = (missing_data / len(df)) * 100
missing_data_summary = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_data_percentage})
print(missing_data_summary)
```

```python
cd=missing_data_summary[missing_data_summary['Percentage']> 50].index
df.drop(columns=cd, inplace=True)
```

```python
df.isna().sum()
```

```python
missing_data = df.isnull().sum()
missing_data_percentage = (missing_data / len(df)) * 100
missing_data_summary = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_data_percentage})
print(missing_data_summary)
```

```python
# Fill missing values for remaining columns
for column in ['manufacturer', 'model', 'condition', 'cylinders', 'fuel', 'title_status', 
               'transmission', 'drive', 'type', 'paint_color']:
    df[column].fillna(df[column].mode()[0], inplace=True)

for column in ['year', 'odometer']:
    df[column].fillna(df[column].median(), inplace=True)
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
x=df.drop(['id','url','region','region_url','description','image_url'],axis=1)
x
```

```python
# Keep only required columns
required_columns = ['price', 'year', 'manufacturer', 'model', 'condition', 'cylinders', 
                    'fuel', 'odometer', 'title_status', 'transmission', 'drive', 
                    'type', 'paint_color']
df = x[required_columns]
```
